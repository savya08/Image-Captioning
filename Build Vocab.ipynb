```
import collections
```

```
def read_captions(path):
  image_to_captions = None
  with open(path) as f:
    image_to_captions = f.read()
  image_to_captions = eval(image_to_captions)
  return image_to_captions
```

```
def build_vocab(image_to_captions, threshold_frequency):
  words = []
  for image_id in image_to_captions.keys():
    caption_list = image_to_captions[image_id]
    for caption in caption_list:
      word_list = caption.split(' ')
      for word in word_list:
        words.append(word)
  counter = collections.Counter(words)
  frequency = dict(counter)
  vocab = []
  for word in frequency.keys():
    if frequency[word] > threshold_frequency:
      vocab.append(word)
  return vocab
```

```
path = '/content/drive/Image Captioning/Utility Data/captions.txt'
image_to_captions = read_captions(path)
vocab = build_vocab(image_to_captions, 10)
print('Vocab size: ', len(vocab))
```

```
with open('/content/drive/Image Captioning/Utility Data/vocabulary.txt', 'w') as f:
  f.write(str(vocab))
```
