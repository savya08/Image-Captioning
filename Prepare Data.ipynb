```
def read_captions(path):
  with open(path) as f:
    captions = f.read()
  captions = captions.split('\n')
  return captions
```

```
def create_mapping():
  image_to_captions = {}
  for c in captions:
    image_id, caption = c.split('\t')
    image_id = image_id.split('.')[0]
    if image_to_captions.get(image_id) == None:
      image_to_captions[image_id] = []
    image_to_captions[image_id].append(caption)
  return image_to_captions
```

```
def clean_data(image_to_captions):
  for image_id, caption_list in image_to_captions.items():
    for i in range(len(caption_list)):
      caption_list[i] = caption_list[i].lower()
  return image_to_captions
```

```
path = '/content/drive/Image Captioning/Dataset/Flickr8k.token.txt'
captions = read_captions(path)[0:-1]
image_to_captions = create_mapping()
image_to_captions = clean_data(image_to_captions)
```

```
with open('/content/drive/Image Captioning/Utility Data/captions.txt', 'w') as f:
  f.write(str(image_to_captions))
```
